```python
import pandas as pd
```

```python
df = pd.read_csv('Meteorite_Landings.csv')
df.head()
```

```python
# Show null values
df.isnull().sum()
```

```python
# Remove null values
df2= df.dropna()
df2.head()
```

```python
df2.isnull().sum()
```

```python
# Rename columns
df3 = df2.rename(columns = {'name': 'meteorite_name', 'mass (g)': 'mass_g', 'year': 'impact_year'})
df3.head()
```

```python
# write formatted to csv
df3.to_csv("Formatted_Meteorite_Data.csv")
```
